Bubble sort

```python
def bubble_sort(arr): #fungsi bubble sort dengan parameter arr
  n = len(arr) #menghitung jumlah elemen dalam array
  for i in range (n-1): #looping
    for j in range (n-i-1): #perulangan untuk mebandingkan elemen yang berjejer
      if arr[j] > arr[j+1]: #mengecek jika elemen kiri lebih besar
        arr[j], arr[j+1] = arr[j+1], arr[j]
  return arr #mengembalikan nilai yang sudah terurut

#contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90] #menyimpan array ke variabel bernama data
sorted_data = bubble_sort(data.copy()) #menggunakan copy agar data asli tidak berubah
print('Data sebelum sorting: ', data) #menampilkan data sebelum diurutkan
print('Data setelah bubble sort: ', sorted_data) #menampilkan data setelah diurutkan
```

Seletion sort

```python
def selection_sort(arr): #fungsi selection sort dengan parameter arr
  n = len(arr) #menghitung jumlah elemen dalam array
  for i in range(n-1): #perulangan untuk menentukan posisi nilai terkecil
    min_index = i
    for j in range (i+1, n): #mencari nilai terkecil dari sisa datanya
      if arr[j] < arr[min_index]: #mengecek jika elemen lebih kecil dari nilai min sementara
        min_index = j #mengganti jika ditemukan nilai yg lebih kecil
      arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

#contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = selection_sort(data.copy()) #data.copy digunakan agar data asli tidak berubah
print('Data sebelum sorting: ', data)
print('Data setelah selection sort: ', sorted_data)
```

Insertion sort

```python
def insertion_sort(arr): #fungsi insertion sort dengan parameter arr
  n = len(arr)
  for i in range(1, n): #looping dari indeks ke-1 hingga terakhir
    key = arr[i] #menyimpan sementara nilai untuk dibandingkan
    j = i-1 #menentukan indeks elemen sebelumnya untuk dibandingkan
    while j >= 0 and arr[j] > key:
      arr[j+1] = arr[j]
      j -= 1
    arr[j+1] = key
  return arr

#contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = insertion_sort(data.copy())
print('Data sebelum soritng: ', data)
print('data setelah insertion sort: ', sorted_data)
```

Merge sort

```python
def merge_sort(arr): #fungsi merge sort dengan parameter arr
  if len(arr) > 1: #jika jumlah elemen array lebih dari 1, maka data akan dibagi
    mid = len(arr) // 2 #menentukan titik tengah
    left_half = arr[:mid] #bagian kiri
    right_half = arr[mid:] #bagian kanan

    merge_sort(left_half)
    merge_sort(right_half)

    i = j = k = 0

    while i < len(left_half) and j < len(right_half): #membandingkan sisi kiri dan kanan
      if left_half[i] < right_half[j]: #bagian kiri (angka yang lebih kecil)
        arr[k] = left_half[i]
        i += 1
      else:
        arr[k] = right_half[j]
        j += 1
        k += 1

      while i < len(left_half): #sisa elemen di bagian kiri(jika ada) dimasukkan ke array utama(k)
        arr[k] = left_half[i]
        i += 1
        k += 1

      while j < len(right_half): #sisa elemen bagian kanan(jika ada) dimasukkan ke array utama(k)
        arr[k] = right_half[j]
        j += 1
        k += 1

  return arr

#contoh penggunan
data = [64, 25, 12, 22, 11]
sorted_data = merge_sort(data.copy())
print('Data sebelum sorting: ', data)
print('Data setelah merge sort: ', sorted_data)
```

Quick sort

```python
def quick_sort(arr): #fungsi quick sort dengan parameter arr
  if len(arr) <= 1: #jika jumlah elemen 1 atau kurang, data langsung dikembalikan karena dianggap sudah terurut
    return arr

  pivot = arr[len(arr) // 2] #menentukan nilai tengah(pivot)
  left = [x for x in arr if x < pivot] #menyimpan nilai elemen yang lebih dari pivot ke bagian kiri
  middle = [x for x in arr if x== pivot] #menyimpan elemen yang bernilai sama dengan pivot ke nilai tengah
  right = [x for x in arr if x > pivot] #menyimpan nilai elemen yang lebih besar dari pivot ke bagian kanan

  return quick_sort(left) + middle + quick_sort(right) #mengurutkan dan menggabungkan

#contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = quick_sort(data.copy())
print('Data sebelum sorting: ', data)
print('Data setelah quick sort: ', sorted_data)
```
