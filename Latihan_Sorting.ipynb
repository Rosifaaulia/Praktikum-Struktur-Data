```python
def bubble_sort(arr): #fungsi bubble sort dengan parameter arr
  n = len(arr) #menghitung jumlah data dalam array
  for i in range (n): #looping
    for j in range (0, n-i-1): #perulangan untuk membandingkan data satu per satu
      if arr[j][1] < arr[j+1][1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
  return arr #mengembalikkan data yang sudah diurutkan

data = [ #list data
    ('Speechless', 4165639),
    ('Popular', 115935470),
    ('Lowkey', 15269600),
    ('Shower', 30297806),
    ('I Like Me Better', 20656985),
    ('Kicau Mania', 3367706)
]

sorted_data = bubble_sort(data.copy())
print('Rekomendasi lagu berdasarkan jmlah pemutaran terbanyak: ')
for item in sorted_data:
  print(item[0], '-', item[1], 'kali diputar') #menampilkan nama lagu dan jumlah pemutaran
```
