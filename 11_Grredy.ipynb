Case 1: Coin Change

```python
def coin_change_greedy(amount, coins):
  coins.sort(reverse=True) #mengurutkan daftar koin dari yang terbesar ke terkecil
  result = [] #list kosong
  for coin in coins: #perulangan untuk setiap koin
    while amount >= coin:
      amount -= coin #mengurangi jumlah uang dengan nilai koin yang dipilih
      result.append(coin) #menambahkan koin yang digunakan ke list
  return result

#contoh penggunaan
amount = 57 #jumlah uang yang ingin ditukar
coins = [25, 10, 5, 1] #koin yang tersedia
change = coin_change_greedy(amount, coins)
print('Koin yang digunakan: ', change)
```

Fractional Knapsack

```python
def fractional_knapsack(items, capacity):
  #items = list of tuples (value, weight)
  items = sorted(items, key=lambda x: x[0]/x[1], reverse=True) #mengurutkan berdasarkan rasio nilai
  total_value = 0.0 #untuk menyimpan total nilai barang yang berhasil dimasukkan
  for value, weight in items: #mengambil barang satu per satu
    if capacity >= weight: #jika kapasitas masih cukup
      capacity -= weight #kapasitas berkurang
      total_value += value #nilai bertambah
    else:
      total_value += value * (capacity / weight)
      break
  return total_value

#contoh penggunaan
items = [(60, 10), (100, 20), (120, 30)] #(value, weight)
capacity = 50
max_value = fractional_knapsack(items, capacity)
print('Nilai maksimum yang dapat dibawa: ', max_value)
```

Minimum Spanning Tree: Prim dan Kruskal

```python
import heapq #membuat priority queue

def prim_mst(graph, start):
  visited = set() #untuk menyimpan kode yang sudah dikunjungi
  min_heap = [(0, start)]
  total_weight = 0 #untuk menjumlahkan seluruh bobot MST

  while min_heap: #selama masih ada node di dalam heap, proses terus berjalan
    weight, node = heapq.heappop(min_heap) #mengambil bobot terkecil
    if node in visited: #cek kunjungan
      continue
    visited.add(node) #tanda sudah dikunjungi
    total_weight += weight #menambahkan bobot
    for neighbor, edge_weight in graph[node]: #memeriksa tetangga A
      if neighbor not in visited:
        heapq.heappush(min_heap, (edge_weight, neighbor))
  return total_weight

#representasi graf: adjacency list
graph = {
    'A': [('B', 2), ('C', 3)],
    'B': [('A', 2), ('C', 1), ('D', 1)],
    'C': [('A', 3), ('B', 1), ('D', 4)],
    'D': [('B', 1), ('C', 4)]
}

print('Total bobot MST (Prim): ', prim_mst(graph, 'A'))
```

```python
def kruskal_mst(edges, n_nodes):
  parent = {i: i for i in range(n_nodes)} #setiap node menjadi 'orang tua' dirinya sendiri

  def find(x): #mencari akar dari suatu node
    while parent[x] != x: #selama parent bukan dirinya sendiri
      x = parent[x] #naik ke parent berikutnya
    return x #mengembalikan root

  def union(x, y): #menggabungkan dua kelompok
    root_x = find(x)
    root_y = find(y)
    if root_x != root_y: #jika root berbeda, maka gabungkan
      parent[root_y] = root_x
      return True
    return False #jika root sama, maka node sudah berada di kelompok yang sama

  edges.sort(key=lambda  x: x[2]) #mengurutkan menurut bobot
  total_weight = 0 #menyimpan total bobot
  for u, v, weight in edges: #memproses edge satu per satu
    if union(u, v):
      total_weight += weight
  return total_weight

#edges: (node1, node2, weight), nodes are represented by integers
edges = [(0, 1, 2), (0, 2, 3), (1, 2, 1), (1, 3, 1), (2, 3, 4)]
print('Total bobot MST (Kruskal): ', kruskal_mst(edges, 4))
```
