Stack dengan list

```python
#python program to demonstrate stack implementation using list
stack = []

#append() function to push element in the stack
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Initial stack')
print (stack)

#pop() function to pop element from stack in LIFO order
print('\nElements popped from stack: ')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack after elements are popped: ')
print(stack)
```

Stack dengan collections.deque

```python
#python program to demonstrate stack implementation using collections.deque
from collections import deque #pendeklarasian variabel

stack = deque()

#append() function to push element in the stack
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Initial stack: ')
print(stack)

#pop( function to pop elemnt from stack in LIFO order)
print('\nElements popped from stack: ')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack after elements are popped: ')
print(stack)
```

Stack dengan queue.LifoQueue

```python
#python program to demonstrate stack immplementation using queue module
from queue import LifoQueue

#Initializing stack
stack = LifoQueue(maxsize=5) #jumlah max item yang diizinkan dalam antrian adalah 5

#qsize() show the number of elements in the stack
print(stack.qsize()) #mengetahui jumlah item sebelum diisi

#put() function to push element in the stack
stack.put('a')
stack.put('i')
stack.put('e')
stack.put('u')
stack.put('o')

print('Full: ', stack.full()) #mengembalikan true jika antrian berisi sesuai jumlah max
print('Size: ', stack.qsize()) #mengembalikan jumlah item sesudah diisi

#get() function to pop element from stack in LIFO order
print('\nElements popped from the stack: ')
print(stack.get())
print(stack.get())
print(stack.get())

print('\nEmpty: ', stack.empty()) #mengembalikkan true jika antrian kosong
```

Stack dengan linked list

```python
class Node:
  def __init__(self, value):
    self.value = value
    self.next = None

class Stack:
  #Initializing a stack, use a dummy node, which is easier for handling edge cases
  def __init__(self):
    self.head = Node('head')
    self.size = 0

  #string representation of the stack
  def __str__(self):
    cur = self.head.next
    out = ''
    while cur:
      out += str(cur.value) + '->'
      cur = cur.next
    return out[:-2]

  #get the current size of the stack
  def getSize(self):
    return self.size

  #check if the stack is empty
  def isEmpty(self):
    return self.size == 0

  #get the top item of the stack
  def peek(self):
    #sanitary check to see if we are peeking an empty stack
    if self.isEmpty():
      raise Exception('Peeking from an empty stack: ')
    return self.head.next.value

  #push a value into the stack
  def push(self, value):
    node = Node(value)
    node.next = self.head.next
    self.head.next = node
    self.size +=1

  #remove a value from the stack an return
  def pop(self):
    if self.isEmpty():
      raise Exception('Pooping from an empty stack')
    remove = self.head.next
    self.head.next = self.head.next.next
    self.size +=1
    return remove.value

#driver code
if __name__ == '__main__':
  stack = Stack()
  for i in range(1, 11):
    stack.push(i) #menambah item sebanyak 10
  print(f'Stack: {stack}')

  for _ in range(1, 6):
    remove = stack.pop() #menghapus item sebanyak 5
    print(f'Pop; {remove}')
  print(f'Stack: {stack}')
```
