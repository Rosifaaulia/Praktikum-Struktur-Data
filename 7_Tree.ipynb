Binary Tree secara manual

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

#Membuat tree secara manual
root = Node(1) #menjadi root
root.left = Node(2) #anak kiri
root.right = Node(3) #anak kanan
root.left.left = Node(4) #anak kiri kedua
root.right.right = Node(5) #anak kanan kedua

#Fungsi inorder traversal (digunakan untuk menelusuri tree dari node terkiri, kemudian induk, lalu anak kanan)
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

print('Hasil inorder traversal dari binary tree: ')
inorder(root) #traversal dimulai dari root
```

Membuat Binary Search Tree (BST)

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

class Binarysearchtree:
  def __init__(self):
    self.root = None #tree kosong

  def insert(self, root, data): #fungsi untuk memasukkan data ke BST
    if root is None:
      return Node(data)
    if data < root.data:
      root.left = self.insert(root.left, data) #jika data lebih kecil dari induk, maka menjadi anak kiri
    else:
      root.right = self.insert(root.right, data) #jika sama/lebih besar, menjadi anak kanan
    return root

bst = Binarysearchtree()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80] #data yang dimasukkan untuk menguji fungsi penyisipan
for value in data_list:
  root = bst.insert(root, value) #perulangan untuk memasukkan data satu per satu
```

Traversal pada BST

```python
def inorder(node): #inorder(terkecil -> terbesar)
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

def preorder(node): #preorder(dari root ke bawah)
  if node:
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

def postorder(node): #postorder(mencetak root setelah semua subtree-nya selesai diproses)
  if node:
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' ')

print('Inorder Traversal: ')
inorder(root)
print('\nPreorder Traversal: ')
preorder(root)
print('\nPostorder Traversal: ')
postorder(root)
```

Pencarian nilai dalam BST

```python
def search(node, key):
  if node is None or node.data == key: #jika node kosong/data ditemukan
    return node #return node tersebut
  if key < node.data: #jika kunci lebih kecil
    return search(node.left, key) #pencarian lanjut ke kiri
  return search(node.right, key) #jika lebih besar, pencarian lanjut ke kanan

#Uji pencarian
key = 60
result = search(root, key)
if result:
  print(f'{key} ditemukan dalam tree.')
else:
  print(f'{key} tidak ditemukan.')

key = 25
result = search(root, key)
if result:
  print(f'{key} ditemukan dalam tree.')
else:
  print(f'{key} tidak ditemukan.')
```
