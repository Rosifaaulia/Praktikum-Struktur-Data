1. Program Struktur Binary Tree Manual

```python
#Struktur Tree berdasarkan identitas saya

class Node:
  def __init__(self, npm):
    self.npm = npm
    self.left = None
    self.right = None

#Binary Tree Manual
root = Node(21)
#21 merupakan jumlah huruf dari nama Rosifa Aulia Qodrun Nada

root.left = Node(11)
#11 merupakan jumlah huruf vokal nama saya

root.right = Node(69)
#69 merupakan 2 digit terakhir npm saya 2505060069

root.left.left = Node(30)
#30 merupakan tanggal lahir saya

root.right.right = Node(4)
#4 merupakan bulan lahir saya

#Menampilkan dalam fungsi inorder
def inorder(node):
  if node:
    inorder(node.left)
    print(node.npm, end=' ')
    inorder(node.right)

print('Hasil inorder traversal dari binary tree: ')
inorder(root)
```

2. Program Binary Search Tree (BST)

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

class Binarysearchtree:
  def __init__(self):
    self.root = None

  def insert(self, root, data):
    if root is None:
      return Node(data)
    if data < root.data:
      root.left = self.insert(root.left, data)
    else:
      root.right = self.insert(root.right, data)
    return root

bst = Binarysearchtree()
root = None
data_list = [69, 30, 4, 82, 79, 27, 90]

#3 digit npm terakhir: 069 dikonversi menjadi integer 69
#tanggal dan bulan lahir: 30 dan 04 (04 dikonversi menjadi integer 4)
#ASCII dua huruf pertama nama R:82 dan O:79
#dua angka random

for value in data_list:
  root = bst.insert(root, value)

def inorder(node): #inorder, left -> root -> right
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

def preorder(node): #preorder, root -> left -> right
  if node:
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

def postorder(node): #postorder, left -> right -> root
  if node:
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' ')

#menampilkan hasil masing-masing traversal
print('---------------------------')
print('Inorder Traversal: ')
inorder(root)
print('\nPreorder Traversal: ')
preorder(root)
print('\nPostorder Traversal: ')
postorder(root)

print('\n---------------------------')
print('HASIL PENCARIAN')
def search(node, key):
  if node is None or node.data == key: ##jika node kosong/data ditemukan
    return node #return node tersebut
  if key < node.data: #jika key lebih kecil
    return search(node.left, key) #masuk ke kiri
  return search(node.right, key) #jika sama / lebih besar, masuk ke kanan

#Uji pencarian
key = 69 #misal key = 69(dua digit npm terakhir)
result = search(root, key)
if result:
  print(f'\n{key} ditemukan dalam tree')
else:
  print(f'{key} tidak ditemukan')

key = 16
result = search(root, key)
if result:
  print(f'\n{key} ditemukan dalam tree')
else:
  print(f'{key} tidak ditemukan')
```
