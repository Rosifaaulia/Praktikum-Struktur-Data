Queue dengan List

```python
#Python program to demonstrate queue implementation using list
queue = []

#append() function to enqueue element in the queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('Initial queue')
print(queue)

#pop(0) function to deque element from queue in FIFO order
print('\nElements dequeued from queue: ')
print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0))

print('\nQueue after removing elements')
print(queue)
```

Queue dengan collections.deque

```python
#python program to demonstrate queue implementation using collections.deque
from collections import deque #pendeklarasian variabel

queue = deque()

#append() function to push element in the stack
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('Initial queue: ')
print(queue)

#pop( function to deque elemnt from queue in FIFO order)
print('\nElements dequed from the queue: ')
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())

print('\nQueue after elements are popped: ')
print(queue)
```

Queue dengan queue.Queue

```python
#Python program to demonstrate queue implementation using queue module
from queue import Queue

#Initializing a queue
qu = Queue(maxsize = 5)

#qsize() show the number of elements in the queue
print(qu.qsize())

#put() function to enqueue element in the queue
qu.put('a')
qu.put('i')
qu.put('u')
qu.put('e')
qu.put('o')

print('Full: ', qu.full())
print('Size: ', qu.qsize())

#get() function to deque element from queue
print('\nElements dequeued from the queue')
print(qu.get())
print(qu.get())
print(qu.get())

print('\nEmpty: ', qu.empty())
print('Full: ', qu.full())
```

Queue dengan single linked list

```python
class Queue:
  def __init__(self):
    self.queue = list()

  def addtoqu(self, dataval):
#insert method to add element
    if dataval not in self.queue:
      self.queue.insert(0, dataval)
      return True
    return False

#pop method to remove element
  def removefromq(self):
    if len(self.queue)>0:
      return self.queue.pop()
    return ('No elements in queue!')

  def size(self):
    return len(self.queue)

TheQueue = Queue()
print(TheQueue.addtoqu('Jan'))
print(TheQueue.addtoqu('Feb'))
print(TheQueue.addtoqu('Mar'))
print(TheQueue.addtoqu('Apr'))
print(TheQueue.size())
print(TheQueue.removefromq())
print(TheQueue.removefromq())
print(TheQueue.size())
```
