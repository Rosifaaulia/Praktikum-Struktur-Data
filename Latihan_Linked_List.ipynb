1. Buatlah double linked list dengan 3 node di dalamnya, kemudian lakukan penyisipan tengah dan hapus dengan kata kunci!

```python
#membuat node
class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None
    self.prevval = None

n1 = Node('Senin') #node head
n2 = Node('Rabu')
n3 = Node('Jumat')

print(n1.dataval)
print(n2.dataval)
print(n3.dataval)
print('======================')

#membuat class double linked list
class DLinkedList:
  def __init__(self):
    self.headval = None

#melakukan penyisipan di tengah
  def AddInbetween(self, mid_node, newdata):
    if mid_node is None:
      print('The mentioned node is absent')
      return
    NewNode = Node(newdata)
    NewNode.nextval = mid_node.nextval
    NewNode.prevval = mid_node
    mid_node.nextval = NewNode

    if NewNode.nextval is not None:
      NewNode.nextval.prevval = NewNode

#menampilakn isi linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

  #menghapus node dengan kata kunci
  def RemoveNode(self, Removekey):
      HeadVal = self.headval
      if (HeadVal is not None):
        if (HeadVal.dataval == Removekey):
          self.headval = HeadVal.nextval
          HeadVal = None
          return
      while (HeadVal is not None):
        if HeadVal.dataval == Removekey:
          break
        prev = HeadVal
        HeadVal = HeadVal.nextval
      if (HeadVal == None):
        return
      prev.nextval = HeadVal.nextval
      HeadVal = None

Li = DLinkedList()
Li.headval = n1 #first node
n1.nextval = n2 #second node
n2.prevval = n1
n2.nextval = n3 #third node
n3.prevval = n2

print('List sebelum penyisipan di tengah:')
Li.listprint()
print('======================')

Li.AddInbetween(n1, 'Selasa')
print('List setelah penyisipan di tengah:')
Li.listprint()
print('======================')

Li.RemoveNode('Jumat')
print('List setelah penghapusan dengan kata kunci: ')
Li.listprint()
```

2. Buatlah circular linked list dengan 3 node di dalamnya, kemudian lakukan penyisipan akhir dan hapus akhir!

```python
#membuat node
class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

n1 = Node('Senin')
n2 = Node('Selasa')
n3 = Node('Rabu')

print(n1.dataval)
print(n2.dataval)
print(n3.dataval)
print('===================')

#membuat class circular linked list
class CLinkedList:
  def __init__(self):
    self.headval = None

#menampilakn isi linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval
      if printval == self.headval:
        break

#melakukan penyisipan di akhir
  def AddInEnd(self, newdata):
    NewNode = Node(newdata)
    if self.headval is None:
      self.headval = NewNode
      NewNode.nextval = self.headval
      return
    last = self.headval
    while(last.nextval is not self.headval):
      last = last.nextval
    last.nextval = NewNode
    NewNode.nextval = self.headval

#menghapus node di akhir
  def RemoveEnd(self):
    if self.headval is None:
      return #jika tidak ada list, maka selesai
    if self.headval.nextval == self.headval: #jika hanya ada 1 node, maka ia menunjuk dirinya sendiri
      self.headval = None #jika dihapus, maka list menjadi kosong
      return
    last = self.headval
    while last.nextval.nextval is not self.headval: #jika lebih dari 1 node, telusuri hingga akhir
      last = last.nextval

    last.nextval = self.headval #node terakhir diarahkan ke head

Li = CLinkedList()
Li.headval = n1 #first node
n1.nextval = n2 #second node
n2.nextval = n3 #third node
n3.nextval = Li.headval #node terakhir kembali ke head

print('List sebelum penyisipan di akhir: ')
Li.listprint()
print('===================')

Li.AddInEnd('Kamis')
print('List setelah penyisipan di akhir: ')
Li.listprint()
print('===================')

Li.RemoveEnd()
print('List setelah dilakukan remove pada node terakhir')
Li.listprint()
```
